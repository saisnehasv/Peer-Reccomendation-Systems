## Feature Engineering Functions

Define functions for extracting comprehensive student features.

```python
def calculate_temporal_features(student_vle, student_id):
    """
    Calculate time-based engagement patterns for a student.
    
    Returns:
        dict: Dictionary containing temporal engagement metrics
    """
    student_data = student_vle[student_vle['id_student'] == student_id]
    if len(student_data) == 0:
        return {
            'early_engagement': 0, 
            'late_engagement': 0,
            'engagement_consistency': 0, 
            'improvement_rate': 0
        }
    
    # Split into early/late course
    mid_week = student_data['week'].max() // 2
    early = student_data[student_data['week'] <= mid_week]['sum_click'].sum()
    late = student_data[student_data['week'] > mid_week]['sum_click'].sum()
    
    # Weekly consistency (inverse of standard deviation)
    weekly_clicks = student_data.groupby('week')['sum_click'].sum()
    consistency = 1 / (1 + weekly_clicks.std()) if len(weekly_clicks) > 1 else 0
    
    return {
        'early_engagement': early,
        'late_engagement': late,
        'engagement_consistency': consistency,
        'improvement_rate': late - early
    }


def calculate_skill_profile(student_assessment, assessments, student_id):
    """
    Extract granular skill profile from different assessment types.
    
    Returns:
        np.array: Skill profile vector [TMA, CMA, Exam, variance, completion_rate]
    """
    student_scores = student_assessment[student_assessment['id_student'] == student_id]
    
    if len(student_scores) == 0:
        return np.zeros(5)
    
    # Merge with assessment metadata
    merged = student_scores.merge(assessments, on='id_assessment', how='left')
    
    # Group by assessment type
    skill_profile = {}
    for assessment_type in ['TMA', 'CMA', 'Exam']:
        type_scores = merged[merged['assessment_type'] == assessment_type]['score']
        skill_profile[assessment_type] = type_scores.mean() if len(type_scores) > 0 else 0
    
    # Add variance and completion rate
    skill_profile['score_variance'] = student_scores['score'].std() if len(student_scores) > 1 else 0
    skill_profile['completion_rate'] = student_scores['score'].notna().mean()
    
    return np.array([
        skill_profile.get('TMA', 0),
        skill_profile.get('CMA', 0),
        skill_profile.get('Exam', 0),
        skill_profile['score_variance'],
        skill_profile['completion_rate']
    ])


print(" Feature engineering functions defined")
print("  - calculate_temporal_features()")
print("  - calculate_skill_profile()")
```

## Parallel Feature Extraction

Build comprehensive student features using parallel processing for efficiency.

```python
def process_student_batch(student_ids, student_info, student_vle, student_assessment, assessments, vle):
    """
    Process a batch of students in parallel to extract all features.
    
    Args:
        student_ids: List of student IDs to process
        
    Returns:
        list: List of feature dictionaries
    """
    feature_list = []
    
    for student_id in student_ids:
        # Basic info
        info = student_info[student_info['id_student'] == student_id].iloc[0]
        
        # Temporal features
        temporal = calculate_temporal_features(student_vle, student_id)
        
        # Skill profile
        skills = calculate_skill_profile(student_assessment, assessments, student_id)
        
        # Overall performance
        student_assess = student_assessment[student_assessment['id_student'] == student_id]
        avg_score = student_assess['score'].mean() if len(student_assess) > 0 else 0
        
        # VLE diversity
        student_vle_data = student_vle[student_vle['id_student'] == student_id]
        if len(student_vle_data) > 0:
            vle_with_type = student_vle_data.merge(vle[['id_site', 'activity_type']], on='id_site', how='left')
            resource_diversity = vle_with_type['activity_type'].nunique()
            total_clicks = student_vle_data['sum_click'].sum()
        else:
            resource_diversity = 0
            total_clicks = 0
        
        # Combine all features
        features = {
            'id_student': student_id,
            'avg_score': avg_score,
            'skill_TMA': skills[0],
            'skill_CMA': skills[1],
            'skill_Exam': skills[2],
            'skill_variance': skills[3],
            'completion_rate': skills[4],
            'early_engagement': temporal['early_engagement'],
            'late_engagement': temporal['late_engagement'],
            'engagement_consistency': temporal['engagement_consistency'],
            'improvement_rate': temporal['improvement_rate'],
            'resource_diversity': resource_diversity,
            'total_clicks': total_clicks,
            'num_prev_attempts': info['num_of_prev_attempts'],
            'studied_credits': info['studied_credits'],
        }
        
        feature_list.append(features)
    
    return feature_list


print("Building comprehensive student features with parallel processing...")
print()

all_students = student_info['id_student'].unique()
n_students = len(all_students)

# Split into batches for parallel processing
batch_size = max(1, n_students // (N_CORES * 4))  # 4x cores for better load balancing
student_batches = [all_students[i:i+batch_size] for i in range(0, n_students, batch_size)]

print(f"Processing {n_students:,} students in {len(student_batches)} batches using {N_CORES} cores...")
print()

# Parallel processing
results = Parallel(n_jobs=N_CORES, verbose=10)(
    delayed(process_student_batch)(
        batch, student_info, student_vle, student_assessment, assessments, vle
    ) for batch in student_batches
)

# Flatten results
feature_list = [item for sublist in results for item in sublist]
student_features_df = pd.DataFrame(feature_list)
student_features_df = student_features_df.fillna(0)

print(f"\n Feature matrix created")
print(f"  Shape: {student_features_df.shape}")
print(f"  Features per student: {student_features_df.shape[1] - 1}")
print(f"\nFeature summary:")
print(student_features_df.describe().round(2))
```

## Pedagogical Complementarity Function

**Key Innovation**: Define complementarity based on educational theory:
1. **Zone of Proximal Development (ZPD)** - optimal skill gap for peer learning
2. **Engagement balance** - similar work ethic prevents imbalance
3. **Learning style diversity** - different approaches benefit both students
4. **Learning pace compatibility** - similar improvement trajectories

```python
def pedagogical_complementarity_score(student_i_features, student_j_features):
    """
    Calculate pedagogical complementarity score between two students.
    
    Based on educational research:
    - Vygotsky's Zone of Proximal Development (ZPD)
    - Peer learning effectiveness research
    - Collaborative learning theory
    
    Args:
        student_i_features: Feature dict for student i
        student_j_features: Feature dict for student j
        
    Returns:
        float: Complementarity score [0, 1] where higher = more complementary
    """
    
    # Extract and validate features
    perf_i = float(student_i_features.get('avg_score', 0))
    perf_j = float(student_j_features.get('avg_score', 0))
    
    eng_i = max(float(student_i_features.get('total_clicks', 0)), 0)
    eng_j = max(float(student_j_features.get('total_clicks', 0)), 0)
    
    # Skill vectors with NaN handling
    skills_i = np.array([student_i_features.get('skill_TMA', 0),
                         student_i_features.get('skill_CMA', 0),
                         student_i_features.get('skill_Exam', 0)], dtype=float)
    skills_j = np.array([student_j_features.get('skill_TMA', 0),
                         student_j_features.get('skill_CMA', 0),
                         student_j_features.get('skill_Exam', 0)], dtype=float)
    
    skills_i = np.nan_to_num(skills_i, nan=0.0)
    skills_j = np.nan_to_num(skills_j, nan=0.0)
    
    # 1. ZPD Score: Optimal gap is around 10-20 points
    perf_gap = abs(perf_i - perf_j)
    zpd_score = np.exp(-((perf_gap - 15)**2) / (2 * 10**2))
    
    # 2. Engagement Balance: Similar work ethic prevents imbalance
    if eng_i == 0 and eng_j == 0:
        eng_score = 0.5
    elif eng_i == 0 or eng_j == 0:
        eng_score = 0.0
    else:
        eng_gap = abs(eng_i - eng_j) / max(eng_i, eng_j)
        eng_score = np.exp(-eng_gap)
    
    # 3. Skill Diversity: Different strengths mean mutual benefit
    skills_i_sum = np.sum(skills_i)
    skills_j_sum = np.sum(skills_j)
    
    if skills_i_sum == 0 and skills_j_sum == 0:
        skill_diversity_score = 0.5
    elif skills_i_sum == 0 or skills_j_sum == 0:
        skill_diversity_score = 0.0
    else:
        try:
            skill_similarity = cosine_similarity([skills_i], [skills_j])[0][0]
            if np.isnan(skill_similarity):
                skill_diversity_score = 0.5
            else:
                skill_diversity_score = 1 - abs(skill_similarity)
        except:
            skill_diversity_score = 0.5
    
    # 4. Learning Pace Compatibility
    pace_i = float(student_i_features.get('improvement_rate', 0))
    pace_j = float(student_j_features.get('improvement_rate', 0))
    pace_gap = abs(pace_i - pace_j)
    pace_score = np.exp(-pace_gap / 1000)
    
    # Weighted combination
    weights = [0.35, 0.30, 0.25, 0.10]
    total_score = (weights[0] * zpd_score + 
                   weights[1] * eng_score +
                   weights[2] * skill_diversity_score +
                   weights[3] * pace_score)
    
    # Validate output
    if np.isnan(total_score) or total_score < 0 or total_score > 1:
        return 0.0
    
    return total_score


# Test the function
sample_i = student_features_df.iloc[0]
sample_j = student_features_df.iloc[100]
comp_score = pedagogical_complementarity_score(sample_i, sample_j)

print(" Pedagogical complementarity function defined")
print(f"\nExample complementarity score: {comp_score:.3f}")
print(f"  Student {int(sample_i['id_student'])}: score={sample_i['avg_score']:.1f}, engagement={sample_i['total_clicks']:.0f}")
print(f"  Student {int(sample_j['id_student'])}: score={sample_j['avg_score']:.1f}, engagement={sample_j['total_clicks']:.0f}")

```

## Training Pair Generation with Class Imbalance Handling

**Critical Issue**: Without careful sampling, we get severe class imbalance:
- Most random pairs have low complementarity (negative class)
- Only ~5-10% of pairs are truly complementary (positive class)

**Solution**: Use stratified sampling with threshold-based labeling

```python
def compute_comp_batch(indices_batch, features_df):
    """
    Compute complementarity for a batch of student pairs.
    
    Args:
        indices_batch: List of (i, j) index tuples
        features_df: DataFrame with student features
        
    Returns:
        list: List of result dictionaries
    """
    results = []
    for i, j in indices_batch:
        student_i = features_df.iloc[i]
        student_j = features_df.iloc[j]
        comp_score = pedagogical_complementarity_score(student_i, student_j)
        results.append({
            'student_i': student_i['id_student'],
            'student_j': student_j['id_student'],
            'comp_score': comp_score
        })
    return results


print("CREATING TRAINING PAIRS WITH CLASS IMBALANCE HANDLING")
print()

MAX_PAIRS = 50000  # Limit for computational efficiency

all_student_ids = student_features_df['id_student'].values
n_students = len(all_student_ids)

# Random sampling of pairs
print(f"Sampling {MAX_PAIRS:,} pairs from {n_students:,} students...")

np.random.seed(RNG)
idx_i = np.random.randint(0, n_students, MAX_PAIRS)
idx_j = np.random.randint(0, n_students, MAX_PAIRS)

# Remove self-pairs
mask = idx_i != idx_j
idx_i = idx_i[mask]
idx_j = idx_j[mask]

print(f"Valid pairs after removing self-pairs: {len(idx_i):,}")
print()

# Create batches for parallel processing
indices = list(zip(idx_i, idx_j))
batch_size = max(1, len(indices) // (N_CORES * 4))
indices_batches = [indices[i:i+batch_size] for i in range(0, len(indices), batch_size)]

print(f"Computing complementarity for {len(indices):,} pairs using {N_CORES} cores...")
print()

# Parallel computation
results = Parallel(n_jobs=N_CORES, verbose=10)(
    delayed(compute_comp_batch)(batch, student_features_df) for batch in indices_batches
)

# Flatten results
pair_data = [item for sublist in results for item in sublist]
pairs_df = pd.DataFrame(pair_data)

print(f"\n Complementarity scores computed for {len(pairs_df):,} pairs")
```

## Class Imbalance Handling: Threshold-Based Labeling

**Strategy**: Use 75th percentile threshold to create balanced classes
- Top 25% of pairs → Positive class (complementary)
- Bottom 75% → Negative class (not complementary)

This maintains a 1:3 ratio which is manageable for most ML algorithms.

## Threshold Sensitivity Analysis


```python
print("Threshold Sensitivity Analysis")
print()

thresholds_to_test = [0.60, 0.70, 0.75, 0.80, 0.85]
print(f"Testing {len(thresholds_to_test)} threshold values:")
print()

for thresh_pct in thresholds_to_test:
    threshold = pairs_df['comp_score'].quantile(thresh_pct)
    temp_labels = (pairs_df['comp_score'] >= threshold).astype(int)
    pos_pct = temp_labels.mean() * 100
    print(f"{int(thresh_pct*100)}th percentile: threshold={threshold:.3f}, positive={pos_pct:.1f}%")

print()
print("Selected threshold: 75th percentile (balanced class distribution)")

```

```python
# Analyze complementarity score distribution
print("\nComplementarity score distribution:")
print(f"  Mean: {pairs_df['comp_score'].mean():.4f}")
print(f"  Std: {pairs_df['comp_score'].std():.4f}")
print(f"  Min: {pairs_df['comp_score'].min():.4f}")
print(f"  25th percentile: {pairs_df['comp_score'].quantile(0.25):.4f}")
print(f"  50th percentile (median): {pairs_df['comp_score'].quantile(0.50):.4f}")
print(f"  75th percentile: {pairs_df['comp_score'].quantile(0.75):.4f}")
print(f"  Max: {pairs_df['comp_score'].max():.4f}")

# Create binary labels using 75th percentile threshold
threshold = pairs_df['comp_score'].quantile(0.75)
pairs_df['label'] = (pairs_df['comp_score'] >= threshold).astype(int)

print(f"\n✓ Binary labels created using threshold: {threshold:.3f}")
print(f"\nClass distribution:")
print(f"  Positive pairs (label=1): {pairs_df['label'].sum():,} ({100*pairs_df['label'].mean():.1f}%)")
print(f"  Negative pairs (label=0): {(1-pairs_df['label']).sum():,} ({100*(1-pairs_df['label']).mean():.1f}%)")
print(f"  Class imbalance ratio: 1:{(1-pairs_df['label']).sum() / pairs_df['label'].sum():.1f}")

# Visualize distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram of complementarity scores
axes[0].hist(pairs_df['comp_score'], bins=50, edgecolor='black', alpha=0.7)
axes[0].axvline(threshold, color='red', linestyle='--', linewidth=2, label=f'Threshold: {threshold:.3f}')
axes[0].set_xlabel('Complementarity Score')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Distribution of Complementarity Scores')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Class distribution bar plot
class_counts = pairs_df['label'].value_counts().sort_index()
axes[1].bar(['Negative (0)', 'Positive (1)'], class_counts.values, color=['lightcoral', 'lightgreen'], edgecolor='black')
axes[1].set_ylabel('Number of Pairs')
axes[1].set_title('Class Distribution After Labeling')
axes[1].grid(True, alpha=0.3, axis='y')

# Add percentage labels on bars
for i, (label, count) in enumerate(class_counts.items()):
    pct = count / len(pairs_df) * 100
    axes[1].text(i, count, f'{count:,}\n({pct:.1f}%)', ha='center', va='bottom')

plt.tight_layout()
plt.savefig(RESULTS_DIR / 'complementarity_distribution.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n Visualization saved to results/complementarity_distribution.png")
```
