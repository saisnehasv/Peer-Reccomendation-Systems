## Create Pair Features

Transform student-level features into pair-level features for classification.

```python
print("Removing score-based features to prevent data leakage...")

# Define features to exclude
score_features = ['avg_score', 'skill_TMA', 'skill_CMA', 'skill_Exam', 
                  'skill_variance', 'completion_rate']

# Filter feature columns
original_features = [col for col in student_features_df.columns if col != 'id_student']
feature_cols = [col for col in original_features if col not in score_features]

print(f"Original features: {len(original_features)}")
print(f"After removing scores: {len(feature_cols)}")
print(f"Removed: {score_features}")
print()

# Then continue with pair feature creation using filtered feature_cols
```

```python
def create_pair_features(student_i_id, student_j_id, features_df):
    """
    Create feature vector for a student pair.
    
    Combines:
    - Absolute differences between features (captures dissimilarity)
    - Element-wise products (captures interactions)
    
    Args:
        student_i_id: ID of first student
        student_j_id: ID of second student
        features_df: DataFrame with student features
        
    Returns:
        np.array: Pair feature vector
    """
    student_i = features_df[features_df['id_student'] == student_i_id].iloc[0]
    student_j = features_df[features_df['id_student'] == student_j_id].iloc[0]
    
    exclude_features = ['avg_score', 'skill_TMA', 'skill_CMA', 'skill_Exam', 
                        'skill_variance', 'completion_rate']
    
    feature_cols = [col for col in features_df.columns 
                    if col != 'id_student' and col not in exclude_features]
    
    # Absolute differences
    diff_features = []
    for col in feature_cols:
        diff_features.append(abs(student_i[col] - student_j[col]))
    
    # Element-wise products (interaction terms)
    product_features = []
    for col in feature_cols:
        product_features.append(student_i[col] * student_j[col])
    
    # Concatenate all features
    return np.concatenate([diff_features, product_features])


def create_pair_features_batch(pairs_batch, features_df):
    """Create pair features for a batch (for parallel processing)."""
    X_batch = []
    y_batch = []
    for _, row in pairs_batch.iterrows():
        pair_feats = create_pair_features(row['student_i'], row['student_j'], features_df)
        X_batch.append(pair_feats)
        y_batch.append(row['label'])
    return X_batch, y_batch


print("Creating pair feature matrices with parallel processing...")
print()

# Split into batches
batch_size = max(1, len(pairs_df) // (N_CORES * 2))
pairs_batches = [pairs_df.iloc[i:i+batch_size] for i in range(0, len(pairs_df), batch_size)]

print(f"Processing {len(pairs_df):,} pairs in {len(pairs_batches)} batches...")

# Parallel processing
results = Parallel(n_jobs=N_CORES, verbose=10)(
    delayed(create_pair_features_batch)(batch, student_features_df) for batch in pairs_batches
)

# Combine results
X_pairs = []
y_labels = []
for X_batch, y_batch in results:
    X_pairs.extend(X_batch)
    y_labels.extend(y_batch)

X_pairs = np.array(X_pairs)
y_labels = np.array(y_labels)

print(f"\n  Pair feature matrix created")
print(f"  Shape: {X_pairs.shape}")
print(f"  Features per pair: {X_pairs.shape[1]}")
```

## Baseline 1: Logistic Regression

**Class Imbalance Handling**: Use `class_weight='balanced'` to automatically adjust weights inversely proportional to class frequencies.

```python
print("BASELINE 1: LOGISTIC REGRESSION WITH CLASS BALANCING")
print()

# Scale features
scaler_lr = StandardScaler()
X_train_scaled = scaler_lr.fit_transform(X_train)
X_test_scaled = scaler_lr.transform(X_test)

# Train with class balancing
print("Training Logistic Regression...")
print("  class_weight='balanced' (adjusts for class imbalance)")
print()

lr_model = LogisticRegression(
    max_iter=1000, 
    random_state=RNG,
    class_weight='balanced',  # KEY: Handles class imbalance
    C=1.0,
    n_jobs=-1
)

lr_model.fit(X_train_scaled, y_train)
print(" Training complete")
print()

# Predictions
y_pred_lr = lr_model.predict(X_test_scaled)
y_prob_lr = lr_model.predict_proba(X_test_scaled)[:, 1]

# Evaluate
lr_metrics = {
    'Accuracy': accuracy_score(y_test, y_pred_lr),
    'Precision': precision_score(y_test, y_pred_lr, zero_division=0),
    'Recall': recall_score(y_test, y_pred_lr, zero_division=0),
    'F1-Score': f1_score(y_test, y_pred_lr, zero_division=0),
    'ROC-AUC': roc_auc_score(y_test, y_prob_lr),
    'PR-AUC': average_precision_score(y_test, y_prob_lr)
}

print("Logistic Regression Results:")
for metric, value in lr_metrics.items():
    print(f"  {metric}: {value:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred_lr, digits=4))

# Confusion matrix
cm_lr = confusion_matrix(y_test, y_pred_lr)
print("\nConfusion Matrix:")
print(f"  TN: {cm_lr[0,0]:,}  FP: {cm_lr[0,1]:,}")
print(f"  FN: {cm_lr[1,0]:,}  TP: {cm_lr[1,1]:,}")
```

## Baseline 2: K-Nearest Neighbors

KNN-based recommendation approach using student feature similarity.

```python
print("BASELINE 2: K-NEAREST NEIGHBORS")
print()

# Prepare student features for KNN
student_feature_cols = [col for col in student_features_df.columns if col != 'id_student']
X_students = student_features_df[student_feature_cols].values

scaler_knn = StandardScaler()
X_students_scaled = scaler_knn.fit_transform(X_students)

# Build KNN model
k = 10
print(f"Training KNN with K={k}...")
knn_model = NearestNeighbors(n_neighbors=k+1, metric='euclidean', n_jobs=-1)
knn_model.fit(X_students_scaled)
print(" Training complete")
print()

# Generate recommendations
print(f"Generating top-{k} recommendations...")
distances, indices = knn_model.kneighbors(X_students_scaled)

all_student_ids = student_features_df['id_student'].values
knn_recommendations = []

for i in range(len(all_student_ids)):
    student_i_id = all_student_ids[i]
    for rank, (idx, dist) in enumerate(zip(indices[i][1:], distances[i][1:]), start=1):
        knn_recommendations.append({
            'student_i': student_i_id,
            'student_j': all_student_ids[idx],
            'rank': rank,
            'distance': dist
        })

knn_recommendations_df = pd.DataFrame(knn_recommendations)

print(f" Generated {len(knn_recommendations_df):,} recommendations")
print(f"  Coverage: {knn_recommendations_df['student_i'].nunique():,} students")

# Evaluate on sample (for computational efficiency)
def eval_knn_batch(batch_df, features_df):
    """Evaluate KNN recommendations using pedagogical complementarity."""
    scores = []
    for _, row in batch_df.iterrows():
        si = features_df[features_df['id_student'] == row['student_i']].iloc[0]
        sj = features_df[features_df['id_student'] == row['student_j']].iloc[0]
        
        # Simple complementarity based on score diff and engagement
        score_diff = abs(si['avg_score'] - sj['avg_score'])
        zpd_score = np.exp(-((score_diff - 15)**2) / (2 * 10**2))
        scores.append(zpd_score)
    return scores

knn_sample = knn_recommendations_df.head(1000)
knn_comp_scores = eval_knn_batch(knn_sample, student_features_df)

knn_metrics = {
    'avg_complementarity': np.mean(knn_comp_scores),
    'std_complementarity': np.std(knn_comp_scores)
}

print(f"\nKNN Quality Metrics (sample of 1000):")
print(f"  Average Complementarity: {knn_metrics['avg_complementarity']:.4f}")
print(f"  Std Complementarity: {knn_metrics['std_complementarity']:.4f}")
```

## Learning Curves


```python
print("Learning Curve Analysis")
print()

from sklearn.model_selection import learning_curve

def analyze_learning_curve(model, X, y, model_name):
    """Generate learning curve to assess model fit."""
    
    train_sizes, train_scores, val_scores = learning_curve(
        model, X, y,
        train_sizes=np.linspace(0.1, 1.0, 10),
        cv=5,
        scoring='roc_auc',
        n_jobs=-1,
        random_state=RNG,
        shuffle=True
    )
    
    train_mean = train_scores.mean(axis=1)
    train_std = train_scores.std(axis=1)
    val_mean = val_scores.mean(axis=1)
    val_std = val_scores.std(axis=1)
    
    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(train_sizes, train_mean, 'o-', label='Training', linewidth=2)
    plt.plot(train_sizes, val_mean, 'o-', label='Validation', linewidth=2)
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.15)
    plt.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, alpha=0.15)
    
    plt.xlabel('Training Set Size', fontsize=11)
    plt.ylabel('ROC-AUC Score', fontsize=11)
    plt.title(f'Learning Curve: {model_name}', fontsize=12)
    plt.legend(fontsize=10)
    plt.grid(True, alpha=0.3)
    plt.ylim([0.7, 0.9]) 
    
    plt.tight_layout()
    plt.savefig(f'results/learning_curve_{model_name.lower().replace(" ", "_")}.png', 
                dpi=150, bbox_inches='tight')
    plt.show()
    
    # Analysis
    gap = train_mean[-1] - val_mean[-1]
    print(f"{model_name}:")
    print(f"  Training: {train_mean[-1]:.4f} (+/- {train_std[-1]:.4f})")
    print(f"  Validation: {val_mean[-1]:.4f} (+/- {val_std[-1]:.4f})")
    print(f"  Gap: {gap:.4f}")
    
    if gap > 0.10:
        print(f"  High variance detected")
    elif gap > 0.05:
        print(f"  Moderate variance")
    elif val_mean[-1] < 0.70:
        print(f"  High bias detected")
    else:
        print(f"  Good fit")
    print()
    
    return val_mean[-1]

# Analyze models
print("Generating learning curves...")
print()

xgb_val = analyze_learning_curve(xgb_model, X_train_scaled, y_train, 'XGBoost')
lr_val = analyze_learning_curve(lr_model, X_train_scaled, y_train, 'Logistic Regression')

print(f"Final validation scores:")
print(f"  XGBoost: {xgb_val:.4f}")
print(f"  Logistic Regression: {lr_val:.4f}")

```

```python
# ROC curves with additional KNN evaluation context
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Left: ROC curves for classifiers
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_prob_lr)
axes[0].plot(fpr_lr, tpr_lr, label=f'Logistic Regression (AUC = {lr_metrics["ROC-AUC"]:.4f})', 
             linewidth=2, color='#2ecc71')

fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_prob_xgb)
axes[0].plot(fpr_xgb, tpr_xgb, label=f'XGBoost (AUC = {xgb_metrics["ROC-AUC"]:.4f})', 
             linewidth=2, color='#3498db')

axes[0].plot([0, 1], [0, 1], 'k--', label='Random Baseline', linewidth=1)

axes[0].set_xlabel('False Positive Rate', fontsize=12)
axes[0].set_ylabel('True Positive Rate', fontsize=12)
axes[0].set_title('ROC Curves - Classification Models', fontsize=14, fontweight='bold')
axes[0].legend(loc='lower right', fontsize=10)
axes[0].grid(True, alpha=0.3)

# Right: KNN complementarity by rank
knn_sample_extended = knn_recommendations_df.head(5000)
complementarity_by_rank = []
for rank in range(1, 11):
    rank_data = knn_sample_extended[knn_sample_extended['rank'] == rank]
    if len(rank_data) > 0:
        scores = []
        for _, row in rank_data.head(100).iterrows():
            si = student_features_df[student_features_df['id_student'] == row['student_i']].iloc[0]
            sj = student_features_df[student_features_df['id_student'] == row['student_j']].iloc[0]
            score_diff = abs(si['avg_score'] - sj['avg_score'])
            zpd_score = np.exp(-((score_diff - 15)**2) / (2 * 10**2))
            scores.append(zpd_score)
        complementarity_by_rank.append(np.mean(scores))
    else:
        complementarity_by_rank.append(0)

ranks = list(range(1, 11))
axes[1].plot(ranks, complementarity_by_rank, marker='o', linewidth=2, 
             markersize=8, color='#e74c3c')
axes[1].set_xlabel('Recommendation Rank', fontsize=12)
axes[1].set_ylabel('Average Complementarity', fontsize=12)
axes[1].set_title('KNN: Complementarity by Recommendation Rank', fontsize=14, fontweight='bold')
axes[1].set_xticks(ranks)
axes[1].grid(True, alpha=0.3)
axes[1].set_ylim([0, max(complementarity_by_rank) * 1.1])

plt.tight_layout()
plt.savefig(RESULTS_DIR / 'roc_and_knn_analysis.png', dpi=150, bbox_inches='tight')
plt.show()

print(" ROC curves and KNN analysis saved to results/roc_and_knn_analysis.png")
```
